```python
#importing the libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# importing the dataset
df = pd.read_table(r"household_power_consumption.txt",sep=';',parse_dates={'Date_Time' : ['Date', 'Time']}, infer_datetime_format=True, 
                 low_memory=False, na_values=['nan','?'],index_col='Date_Time')
df
```

```python
df.shape
```

```python
# basic info about the dataset
df.info()
```

```python
df.head()
```

```python
df.tail()
```

```python
# to get the count of null values


df.isnull().sum()
```

```python
 # fill missing values row wise and making the changes permanent in the original dataframe

df.ffill(axis=0,inplace=True)  
```

```python
df.isnull().sum()
```

```python
df['Power_consumption'] = (df['Global_active_power']*1000/60)-df['Sub_metering_1']-df['Sub_metering_2']-df['Sub_metering_3']
```

```python
df
```

```python
#With the help of new column 'Date' & 'Time',it will be easier to do grouping on the data which willl ease the work of Visualization for better understanding on Data
df['Date'] = df.index.date    
df['Time'] = df.index.time
```

```python
df
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df.info()
```

```python
df = df[df.index.year>2006]
```

```python
df.info()
```

```python
New_df = df.groupby(['Date']).sum()
```

```python
New_df
```

# Creating all three Sub-DataFrame from original dataframe

```python
#power consumption DataFrame(watt-hrs)

New_df_power_consumption = New_df[['Power_consumption']] 
```

```python
#Yearly Statistical analysis of power consumption

Yeary_power_consumption=New_df_power_consumption.groupby(New_df_power_consumption.index.year).describe()
```

```python
New_df_power_consumption
```

```python
#Creating sub-dataframe of sub-metering

New_df_sub_meterings = New_df[['Sub_metering_1','Sub_metering_2','Sub_metering_3']] 
```

```python
New_df_sub_meterings
```

```python
# Creating sub-dataframe of Global_active_power, Global_reactive_power, and Global_intensity

New_df_active_reactive =New_df[['Global_active_power','Global_reactive_power','Global_intensity']] 
```
